```python
total_spend = 0
for ex in expenses :
  total_spend += ex["amount"]
print(f"Total Expenses : {total_spend}")
```

```python
category = {}
for expense in expenses :
  cate = expense['category'].title()
  if cate not in category:
    category[cate] = expense['amount']
  else :
    category[cate] += expense['amount']


max_value = max(category.values())



cat={}
for ex in expenses :
  amount = ex['amount']
  name = ex['name']
  if name not in cat:
    cat[name] = amount
  else :
    cat[name] += amount

max_spend_by_category = max(cat.values())


```

```python
def print_reports():
  for i , j in category.items() :
    print(f"{i} : {j}\n")
  print (50 * "-")
  for k, v in category.items():
    if v == max_value:
        print(f"Highest category: {k} ({v})\n")
  print (50 * "-")
  for i,j  in cat.items():
    if j == max_spend_by_category:
     print(f"Highest category: {i} ({j})\n")
```

```python
print_reports()
```
