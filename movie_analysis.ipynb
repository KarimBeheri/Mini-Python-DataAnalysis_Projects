```python
#Highest Rated Movie
for Highest_Rated in range(len(movies)):

  Highest_Rated= max(movies, key=lambda x: x['rating'])
print(f"Highest Rated Movie : {Highest_Rated['name']} ({Highest_Rated['rating']})")
```

```python
for Lowest_Rated in range(len(movies)):
  Lowest_Rated = min(movies , key=lambda movie : movie['rating'])
print(f"Lowest Rated Movie : {Lowest_Rated['name']} ({Lowest_Rated['rating']})")
```

```python
genre_totals = {}
genre_counts = {}
for movie in movies:
    genre = movie["genre"]
    rating = movie["rating"]
    if genre in genre_totals :
      genre_totals[genre] += rating
      genre_counts[genre] += 1
    elif genre not in genre_totals:
      genre_totals[genre] =rating
      genre_counts[genre] = 1

print("Average Ratings by Genre:")
for genre in genre_totals:
    average = genre_totals[genre] / genre_counts[genre]
    print(f"{genre}: {average:.2f}")

print(40 * "*")

for genre in genre_totals:
    print(f"{genre}: {genre_counts[genre]}")
```
